# Predict if customer will need a housing loan or not

```python
import numpy as np
import pandas as pd


newDF = pd.read_csv('bank-full-dataset.csv')
newDF
```

### Cleaning the dataset

```python
#change prediction column name

newDF.rename(columns={'y':'ifCreatedAccount'},inplace=True)
newDF.rename(columns={'housing':'housingLoan'}, inplace =True)

newDF
```

```python
#drop columns with too many 'unknown' values
#drop contact and poutcome

newDF.drop(['contact', 'poutcome'], axis='columns', inplace=True)

##after doing filter based feature selection in Azure Machine Learning Studio, the highest 7 features were selected
#drop columns that aren't required

newDF.drop([ 'previous', 'duration', 'default', 'marital', 'loan', 'day', 'campaign'], axis='columns', inplace=True)
newDF
```

```python
newDF['education'].value_counts()
```

```python
#change yes = 1 no = 0
#give values to each job and each month from 1 to 12

newDF['ifCreatedAccount'] = newDF['ifCreatedAccount'].map({'yes': 1, 'no': 0})
newDF['month'] = newDF['month'].map({'jan': 1, 'feb': 2,'mar':3,'apr': 4,'may': 5, 'jun': 6, 'jul': 7, 'aug': 8, 'sep': 9,'oct':10,'nov': 11,'dec': 12})
newDF['housingLoan'] = newDF['housingLoan'].map({'yes': 1, 'no': 0})
newDF['job'] = newDF['job'].map({'blue-collar': 1, 'management': 2,'technician':3,'admin.': 4,'services': 5, 'retired': 6, 'self-employed': 7, 'entrepreneur': 8, 'unemployed': 9,'housemaid':10,'student': 11,'unknown': 12})
newDF['education'] = newDF['education'].map({'primary': 1, 'secondary': 2,'tertiary':3,'unknown': 0})


newDF
```

```python
#rearrange data columns

newDF = newDF[['age','job', 'education', 'balance', 'month', 'pdays', 'ifCreatedAccount', 'housingLoan']]
newDF.head()
```

```python
newDF['housingLoan'].value_counts()
```

```python
# Plot of Count
import matplotlib.pyplot as plt

newDF['housingLoan'].value_counts().plot(kind='bar')
plt.ylabel('Count')
plt.xlabel('Has housing loan or not')
```

```python
#final dataset after cleaning

newDF . to_csv("cleanedHL.csv", index=False)
```

```python
df2 = pd.read_csv('cleanedHL.csv')
df2
```

### Creating models

```python

### Independent and Dependent features



X=df2.iloc[:,:-1]
y=df2.iloc[:,-1]
```

```python
X.head()
```

```python
y.tail()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# Implement Random Forest classifier
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
# Prediction
y_pred=rfc.predict(X_test)
```

```python
# Check Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```

```python
# Create a Pickle file using serialization for Random Forest Classifier
import pickle
pickle_out = open("H-RandomForestC.pkl","wb")
pickle.dump(rfc, pickle_out)
pickle_out.close()
```

```python
#Implement Gaussian Naive Bayes classifier

from sklearn.naive_bayes import GaussianNB

NBclassifier = GaussianNB()
NBclassifier.fit(X_train,y_train)
```

```python
NBy_pred=NBclassifier.predict(X_test)
```

```python
#checking accuracy
NBscore=accuracy_score(y_test,NBy_pred)
NBscore
```

```python
#Implement Decision Tree classifier

from sklearn.tree import DecisionTreeClassifier

DTclassifier = DecisionTreeClassifier()
DTclassifier.fit(X_train,y_train)
```

```python
DTy_pred=DTclassifier.predict(X_test)
```

```python
#checking accuracy
DTscore=accuracy_score(y_test,DTy_pred)
DTscore
```

```python
DT_pickle_out = open("H-DTClassifier.pkl","wb")
pickle.dump(DTclassifier, DT_pickle_out)
DT_pickle_out.close()
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier

KNNclassifier = KNeighborsClassifier(n_neighbors=5)
KNNclassifier.fit(X_train,y_train)
```

```python
KNNy_pred=KNNclassifier.predict(X_test)
```

```python
#checking accuracy
KNNscore=accuracy_score(y_test,KNNy_pred)
KNNscore
```

```python
KNN_pickle_out = open("H-KNNClassifier.pkl","wb")
pickle.dump(KNNclassifier, KNN_pickle_out)
KNN_pickle_out.close()
```
